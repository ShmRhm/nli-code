```python
import numpy as np
import matplotlib.pyplot as plt

# Data 
languages = ['ZHO', 'SPA', 'FRA', 'ITA', 'TUR', 'DEU']
topics = ['Topic 1', 'Topic 2', 'Topic 3', 'Topic 4', 'Topic 5']
f1_scores = [
    [0.77, None, 0.86, 0.80, 0.80],  # ZHO
    [0.50, 0.61, None, None, None],  # SPA
    [0.88, 0.90, None, None, None],  # FRA
    [0.78, 0.56, None, None, None],  # ITA
    [0.98, None, None, None, None],  # TUR
    [None, 0.00, None, None, None]   # DEU
]
support = [
    [40, None, 156, 149, 94],  # ZHO
    [18, 32, None, None, None],  # SPA
    [14, 44, None, None, None],  # FRA
    [11, 23, None, None, None],  # ITA
    [57, None, None, None, None],  # TUR
    [None, 1, None, None, None]   # DEU
]

# Function to create radar chart
def create_radar_chart(language, f1_scores, support):
    categories = topics
    N = len(categories)
    
    # Convert None to 0 for missing data
    f1_scores = [score if score is not None else 0 for score in f1_scores]
    support = [score if score is not None else 0 for score in support]
    
    # Compute angle for each axis
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()

    # Create the plot
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    
    # Plot F1 scores
    ax.fill(angles, f1_scores, color='blue', alpha=0.25)
    ax.plot(angles, f1_scores, color='blue', linewidth=2, label=f'{language} F1')

    # Plot Support scores on a second layer (with lighter alpha for visibility)
    ax.fill(angles, support, color='green', alpha=0.25)
    ax.plot(angles, support, color='green', linewidth=2, label=f'{language} Sup')

    # Add labels and title
    ax.set_yticklabels([])
    ax.set_xticks(angles)
    ax.set_xticklabels(categories, fontsize=12)
    ax.set_title(f"Radar Chart for {language}", fontsize=14)

    plt.show()

# Create radar chart for each language
for i, language in enumerate(languages):
    create_radar_chart(language, f1_scores[i], support[i])
```
